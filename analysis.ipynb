```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
runs=pd.read_csv("out.txt").values.reshape(-1)
```

```python
hist=pd.DataFrame(np.unique(runs, return_counts=True)).T.set_index(0)[1]
hist/=hist.sum()/100
hist.plot(title='distrib',xlabel='shots',ylabel='p (%)',grid=True)
```

```python
mean_evolution=np.cumsum(runs)/np.arange(runs.size)
pd.Series(mean_evolution[500:]).plot(title='mean_evolution',xlabel='occurencies',ylabel='mean',grid=True)
```

```python
p40_evolution=np.cumsum(runs==40)/np.arange(runs.size)*100
pd.Series(p40_evolution[500:]).plot(title='p40_evolution',xlabel='occurencies',ylabel='mean',grid=True)
```

```python
p0_evolution=np.cumsum(runs==0)/np.arange(runs.size)*100
pd.Series(p0_evolution[500:]).plot(title='p0_evolution',xlabel='occurencies',ylabel='mean',grid=True)
```
